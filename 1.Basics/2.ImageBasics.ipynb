```python
import cv2
import numpy as np
img = np.zeros((512,512,3), np.int8)
drawing = False
ix = -1
iy = -1


def drw_rect(event,x,y,flags,params):
    global ix, iy,drawing
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
    elif event == cv2.EVENT_MOUSEMOVE:
        cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1) #-1 here means solid circle
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.rectangle(img, (ix,iy),(x,y),(0,255,0),-1)
cv2.namedWindow(winname = 'my_drw')
cv2.setMouseCallback('my_drw', drw_rect)
while True:
    cv2.imshow('my_drw',img)
    if cv2.waitKey(1) & 0xFF == 27:
        break
```
